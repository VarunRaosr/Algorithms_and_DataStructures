```python
class TreeNode:
    def __init__(self,data):
        self.key = data
        self.left = None
        self.right = None
        self.parent = None

def inOrder(root):
    if root is not None:
        inOrder(root.left)
        print(root.key)
        inOrder(root.right)

def get_tree(root,stck):
    if root is not None:
        get_tree(root.left,stck)
        stck.append(root)
        get_tree(root.right,stck)
        return stck
    
def fixBST(root):
    stck = []
    stck = get_tree(root,stck)
    first = None
    last = None
    
    for index in range(1,len(stck)):
        if stck[index-1].key > stck[index].key:
            if first is None:
                first = stck[index-1]
            else:
                last = stck[index]
    if first is None:
        print("Tree is sorted")
        return
    if last is None:
        last = stck[-1]
    # found elements now exchange them    
    fix_elements(root,first,last)
    return

def fix_elements(root,first,last):
    temp = first.key
    first.key = last.key
    last.key = temp
    return

        
def insert(root,data):
    if root is None:
        root = TreeNode(data)
    else:
        if data < root.key:
            if root.left is None:
                root.left = TreeNode(data)
                root.left.parent = root
                return
            else:
                insert(root.left,data)
        else:
            if root.right is None:
                root.right = TreeNode(data)
                root.right.parent = root
                return
            else:
                insert(root.right,data)
    
r = TreeNode(10)
r.left = TreeNode(15)
r.right = TreeNode(5)

r.left.parent = r
r.left.parent = r

r.left.left = TreeNode(4)
r.left.right = TreeNode(7)

r.left.left.parent = r.left
r.left.right.parent = r.left

r.right.left = TreeNode(14)
r.right.right = TreeNode(17)

r.right.left.parent = r.right
r.right.right.parent = r.right

print("Before".center(40,'-'))
inOrder(r)
fixBST(r)
print("After".center(40,'-'))
inOrder(r)
```
